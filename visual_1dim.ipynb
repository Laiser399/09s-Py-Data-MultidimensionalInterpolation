```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.pyplot import figure
from lib import get_x, build_poly, build_poly_adaptively, make_evaluator

def fun(arg):
    return np.sin(arg * 8)
```

```python
a_coefficients = build_poly(fun, 3)
arg_table = np.linspace(0, 1, 100)
base_table = fun(arg_table)
evaluator = make_evaluator(a_coefficients)
poly_table = np.array([evaluator(arg) for arg in arg_table])
figure(figsize=(12, 10))
plt.plot(arg_table, base_table, arg_table, poly_table);
```

```python
a_coefficients = build_poly_adaptively(fun, 1e-3)
arg_table = np.array([get_x(l, i) for (l, i) in a_coefficients.keys()])
evaluator = make_evaluator(a_coefficients)
poly_table = np.array([evaluator(arg) for arg in arg_table])
figure(figsize=(12, 10))
plt.plot(arg_table, poly_table, '.', )
plt.plot(arg_table, np.zeros(len(arg_table)), '.');
```

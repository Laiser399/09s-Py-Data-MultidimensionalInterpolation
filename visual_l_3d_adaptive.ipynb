```python
from functions import functions_3d, functions_3d_names
from interpolator import get_adaptive_indices_iterator, InterpolationMode
from visual_common import calc_errors
from visual_common import display_3d_grid
from visual_common import get_points_counts
from visual_common import interpolate_table

# убрал последнюю из-за сложности вычислений (на err = 0.1 кол-во узлов доходило до 40000)
functions_3d = functions_3d[:-1]
functions_3d_names = functions_3d_names[:-1]
```

# Интерполирование 3-мерных функций на адаптивной сетке на линейном базисе

```python
dim = 3
errors = [10 ** degree for degree in range(-1, -4, -1)]
indices_iterators = [get_adaptive_indices_iterator(err) for err in errors]

interpolated = interpolate_table(functions_3d, dim, InterpolationMode.Linear, indices_iterators)
```

## Сетки

```python
display_3d_grid(interpolated)
```

## Кол-во узлов

```python
row_names = [f'err: {err}' for err in errors]
get_points_counts(interpolated, row_names, functions_3d_names)
```

## Погрешности

```python
calc_errors(interpolated, functions_3d, dim, row_names, functions_3d_names)
```

# Интерполирование 2-мерных функций на адаптивной сетке на линейном базисе

## Сетки

```python
row_names = [f'err: {err}' for err in errors]
display_2d_grid(interpolated, row_names, functions_2d_names)
```

## Кол-во узлов

```python
get_points_counts(interpolated, row_names, functions_2d_names)
```

## Погрешности

```python
calc_errors(interpolated, functions_2d, dim, row_names, functions_2d_names)
```
